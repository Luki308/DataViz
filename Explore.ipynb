```python
import os

import clustbench
import genieclust

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
data_path = os.path.join("clustering-data-v1")
data_path = os.path.abspath(data_path)
data_path
```

# Test battery

```python
# Print the battery names
print(f"Battery names: {clustbench.get_battery_names(path=data_path)}")

# Print the dataset names for a specific battery
battery = "wut"
print(f"Dataset names for battery '{battery}': {clustbench.get_dataset_names(battery, path=data_path)}")

# Load a specific dataset from a battery and print its details
dataset = "x2"
b = clustbench.load_dataset(battery, dataset, path=data_path)
# Print the dataset details
print(f"Dataset details: {b}")
print(f"Dataset shape: {b.data.shape}")
```

## Plot one of the datasets

```python
for i in range(len(b.labels)):
    plt.subplot(1, len(b.labels), i + 1)
    genieclust.plots.plot_scatter(
        b.data,
        labels=b.labels[i]-1,
        axis='equal',
        title=f"Labels {i}",
    )
plt.show()
```

# Fetching precomputed results

```python
# results_path = None # nie jest pobrane
# method_group = "*"  # or "*" for everything
# res = clustbench.load_results(
#     method_group, b.battery, b.dataset, b.n_clusters, path=results_path
# )
# print(list(res.keys()))
```

```python
kmeans_clusterer = KMeans(random_state=0, n_init=10)
res = clustbench.fit_predict_many(
    kmeans_clusterer,
    b.data,
    b.n_clusters,
)
print(f"Results: {res}")
clustbench.get_score(b.labels, res)
```

```python
for i, k in enumerate(res.keys()):
    print(f"Labels {i}: {k}")
    # print(f"Labels {i}: {res[i]}")
    plt.subplot(1, len(res), i + 1)
    genieclust.plots.plot_scatter(
        b.data,
        labels=res[k]-1,
        axis='equal',
        title=f"Kmeans Labels (k = {k}) ",
    )
```
